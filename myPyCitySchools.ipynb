```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# These are constants
SCHOOL_DATA_PATH = Path("Resources/schools_complete.csv")
STUDENT_DATA_PATH = Path("Resources/students_complete.csv")

MIN_PASSING_SCORE = 70
```

```python
school_df = pd.read_csv(SCHOOL_DATA_PATH)
school_df
```

```python
student_df = pd.read_csv(STUDENT_DATA_PATH)
student_df
```

```python
# Combine the data into a single dataset.   # Merge the two data frames: school_df and student_df in a new df called school_data_complete
school_data_complete = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"]) 
print(school_data_complete.shape) # one print per cell block; make it .shape instead of .head to get the total number of rows and columns displayed first. .head only gives the first 5 as default
school_data_complete
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_complete["school_name"].nunique()
print(school_count)
school_count
```

```python

#Total number of students.  Should be the same as the total number of rows for Student ID. 
student_count = school_data_complete["student_name"].count()
print(student_count)
```

```python
#Summary of numberic series in df.  Curious to see / test.  
school_data_complete.describe()
```

```python
# Calculate the total budget. worked with using the school_df the school_df
total_budget = school_df["budget"].sum()
print(total_budget)
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
print(average_math_score)
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = student_df["reading_score"].mean()
print(average_reading_score)
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= MIN_PASSING_SCORE )].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >=70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100 
passing_reading_percentage 
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
# Formatting
district_summary = pd.DataFrame()
district_summary["Total Schools"] = [school_count]
district_summary["Total Students"] = [student_count]
district_summary["Total Budget"] = [total_budget]
district_summary["Average Math Score"] = [average_math_score]
district_summary["Average Reading Score"] = [average_reading_score]
district_summary["% Passing Math"] = [passing_math_percentage]
district_summary["% Passing Reading"] = [passing_reading_percentage]
district_summary["% Overall Passing"] = [overall_passing_rate]

#district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
#district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete["school_name"].value_counts()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
school_budget_summaries = pd.DataFrame()
print(school_data_complete.columns)
for school, school_gdf in school_data_complete.groupby("school_name"):     
    a_single_school_budget = school_gdf['budget'].max() 
    a_single_School_Type =  school_gdf['type'].iloc[0]                                       
    single_school_student_count = school_gdf['Student ID'].nunique()
    single_school_spending_per_capita = a_single_school_budget/single_school_student_count
    school_budget_summaries.loc[school, "Total School Budget"] = a_single_school_budget
    school_budget_summaries.loc[school, "Total Students"] = single_school_student_count
    school_budget_summaries.loc[school, "Per Student Budget"] = single_school_spending_per_capita
    school_budget_summaries.loc[school, "School Type"] = a_single_School_Type

print (school_budget_summaries)
```

```python
tmpdf2 = school_data_complete.copy(deep=True)
tmpdf2["passing_math"] = tmpdf2['math_score']>=MIN_PASSING_SCORE
tmpdf2["passing_reading"] = tmpdf2['reading_score']>=MIN_PASSING_SCORE
tmpdf2["passing_reading_and_math"] = tmpdf2["passing_math"] & tmpdf2["passing_reading"]
school_passing_data = tmpdf2[["passing_math","passing_reading","passing_reading_and_math",'school_name']].groupby('school_name').agg('mean')
school_passing_data = school_passing_data.applymap(lambda x: x*100)
school_passing_data = school_passing_data.rename(columns={"passing_math":"% Passing Math", 'passing_reading':"% Passing Reading", 'passing_reading_and_math':"% Overall Passing"})
print(school_passing_data)
print(school_passing_data.columns)
```

## Scores by School Size

```python
per_school_summary_df
```

```python
per_school_summary_df.rename(columns = {('reading_score','mean'):"Average Reading Score"}, inplace=True)
per_school_summary_df.rename(columns = {('math_score','mean'):"Average Math Score"}, inplace=True)
per_school_summary_df
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary_df.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary_df.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary_df.groupby(["School Size"])["% Overall Passing"].mean()
```
